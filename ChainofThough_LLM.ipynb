Understanding Chain of Thought with Backtracking

The core idea behind "Chain of Thought with Backtracking" is to help the LLM think backward after analyzing the situation instead of being only a one-way train. As you might remember, LLMs are designed to be next-word-prediction machines, and in case problems require a lookback, they naturally tend to become less and less effective. Our goal as Prompt Engineers is to help them navigate this challenge by creating specific instructions on how to backtrack after getting to a certain point.
Simple Example

Let's take a look at a simple example prompt that includes Chain of Thought and nothing more.

Markdown

__ASK__


Find a 10-digit number where the first digit is how many zeros in the number, the second digit is how many 1s in the number etc. until the tenth digit which is how many 9s in the number.


__REQUIREMENTS__

- Let's think Step-by-Step to solve this problem

Sample output:

Step 1: 

The number has 10 digits. Let's assume the number is represented as abcdefghij. Now we need to find the value of each digit from 'a' to 'j' based on the given conditions.


Step 2:

The first digit is how many zeros are in the number. Since there are no zeros in the number, the first digit must be 0.


Step 3:

The second digit is how many ones are in the number. To have one one in the number, the second digit must be 1.


Step 4:

Continuing this pattern, the third digit must be 2, the fourth digit must be 3, and so on until the tenth digit is 9.


Step 5:

Putting all the digits together, the number we are looking for is 0123456789.

As you can see the LLM gives it a good attempt but gets the result completely wrong. The reason is that this is the kind of problem where working backward from the end is much easier than the other way around.
Introducing Backtracking

While progressing linearly through a problem is effective by using the standard Chain Of Thought method, there are instances where you might need to revisit and revise earlier steps. This is where "Backtracking" comes into play. It involves going back after the initial analysis to come up with the final answer. Let's enhance the prompt above to see an example:

Markdown

__ASK__


Find a 10-digit number where the first digit is how many zeros in the number, the second digit is how many 1s in the number etc. until the tenth digit which is how many 9s in the number.


__REQUIREMENTS__


- Forward Analysis

  - Think step-by-step to analyze the problem - do not suggest an answer.

- Backtracking

   - After the initial step-by-step analysis, work backward from the end to come up with the answer.

Notice here a key component do not suggest an answer - this is done to avoid biasing the LLM. If the LLM produces an answer while doing the forward analysis (aka Chain of Thought), it will be more likely to stick to that answer even after Backtracking, even though it's likely to be incorrect.

Sample output:

1. Let's start by listing out the requirements:

   - The first digit is how many zeros are in the number

   - The second digit is how many ones are in the number

   - Continue this pattern up to the tenth digit, which is how many nines are in the number


2. Think about the possible constraints:

   - The total number of digits is 10

   - Each digit can range from 0 to 9


3. Let's consider the problem in reverse:

   - The tenth digit has to be the number of nines in the number, so we already know it will be 0 since there are no nines in the number yet.

   - The ninth digit will be the number of eights in the number, which is also 0 so that the ninth digit will be 0 as well.

   - Continue this pattern up to the first digit, which will be the number of zeros in the number.


4. Work backward from the last digit:

   - The tenth digit is 0

   - The ninth digit is 0

   - The eighth digit is 0

   - The seventh digit is 1

   - The sixth digit is 0

   - The fifth digit is 0

   - The fourth digit is 0

   - The third digit is 1

   - The second digit is 2

   - The first digit is 6


Therefore, the 10-digit number that fits the criteria is 6210001000.

This time the answer is absolutely correct and you can see the overall depth of the analysis is much stronger.
Other Considerations

In solving complex problems like the one above and in dealing with complicated scenarios applying one technique like Chain of Thought with Backtracking won't be enough. Even in this scenario, if you were to run the prompt above multiple times you'd see that not very advanced LLMs don't always get the right answer.

This is where the combination of advanced techniques comes to the rescue. You might remember the method of Brainstorming that we've learned in this course. Instead of just running the above prompt once and trusting the output, you could run it 5-10 times and then write a consolidator prompt that reviews all the outputs and determines which solution is correct. That will dramatically increase the ability of your prompt to get to the right answer.

Sure, let's solve this problem using both forward analysis and backtracking.

### Forward Analysis:
1. **Understanding the Problem**: We are looking for a 10-digit number where each digit represents the count of occurrences of the digit corresponding to its position. For example, the first digit tells us how many zeros are in the number, the second digit tells us how many ones are present, and so on until the tenth digit tells us how many nines are present.

2. **Initialization**: Start with a structure `a`, an array of 10 elements, `a[0], a[1], ..., a[9]` where `a[i]` represents how many times digit `i` appears in the number we are trying to find.

3. **Constraints Observation**:
   - The sum of the array `a` should be 10 because the number is exactly 10 digits long.
   - For instance, `a[0]` (count of zeros) must capture how many zeros are present, but having too many zeros will conflict with the total sum of 10 rule.

4. **Exploring Possibilities**:
   - Start by considering edge cases. If `a[0] = 9`, then there must be a single non-zero digit somewhere to make the sum 10, which doesn't represent a valid number.
   - Continue evaluating what it means if `a[x] = y` in terms of how to set up the remaining digits.

5. **Data-Driven Insights**:
   - A number where each `a[i]` satisfies its role and sums precisely to 10, typically involves some nested calculations and satisfaction of constraints.

### Backtracking:
1. **Start from the Known Requirement (Sums to 10)**:
   - Assume we already have `a[i]` setup such that they potentially could make a valid number.
   - Consider one possible assignment based on thoroughly calculated guesses or known methodologies, such as other solveable constraints in similar puzzles: `a = [6, 2, 1, 0, 0, 0, 0, 0, 0, 1]`.

2. **Work Backwards to Validate**:
   - Count the Calculated Numbers: With `a = [6, 2, 1, 0, 0, 0, 0, 0, 0, 1]`, verify each assignment:
     - Six 0s imply `a[0]` can be 6.
     - Two 1s imply `a[1]` can be 2.
     - One 2 implies `a[2]` can be 1.
     - Finally, one 9 means `a[9]` should be 1.

3. **Validation**:
   - Confirm the entire setup: `6210001000` indeed meets all specified criteria since:
     - There are six zeros,
     - Two ones,
     - One two,
     - One nine completing the setup consistent with our start.
   - Each requirement satisfies its position: the first digit is the count of zeros; the sixth digit reflects the single occurrence of a nine, etc.

Thus, the only valid solution to this self-descriptive number problem is indeed `6210001000`.

__ASK__
Find a 10-digit number where the first digit is how many zeros in the number, the second digit is how many 1s in the number etc. until the tenth digit which is how many 9s in the number.

__REQUIREMENTS__
- Brainstorming
   - Before moving forward with solving the problem, develop 3 approaches to solve this problem without giving a solution.
- Forward Analysis
  - Think step-by-step to analyze the problem - do not suggest an answer.
- Backtracking
   - After the initial step-by-step analysis, work backward from the end to come up with the answer.

---

__BRAINSTORMING__

1. **Iterative Approach**: Start with an arbitrary 10-digit number and iterate through permutations, checking each one to see if it meets the criteria. Although this could be computationally expensive, it systematically checks all possibilities.

2. **Constraint Satisfaction Method**: Model this as a constraint satisfaction problem (CSP) where each digit represents the count of a particular digit. Set up equations for each digit and use constraint-solving techniques to find the number.

3. **Simulation of Digit Counts**: Create a simulation that updates digit counts iteratively. Begin with an assumption for how many digits each could have, adjust based on inconsistencies in the logical counts, and try converging to a stable solution.

__FORWARD ANALYSIS__

1. **Understand Constraints**: Each digit represents how many times 0 through 9 appear in the number. For example, the first digit tells us the quantity of 0s, the second the quantity of 1s, and so on. The sum of all digits should equal 10, as we have a 10-digit number.

2. **Set Up Equations**: We must consider individual equations for each digit. If something like x_i represents the number at position i, then:
   - \( x_0 + x_1 + ... + x_9 = 10 \)
   - Each \( x_i \) must accurately reflect the count of that digit within the number.

3. **Digit Bounds and Initial Assumptions**: Digit counts generally fall between 0 and 9. Start thinking about possibilities where there are more frequent occurrences of smaller digits, especially as having 9s or 8s too frequent will not satisfy constraints.

__BACKTRACKING__

1. **Validation of End State**: Assume hypothetical solutions by filling in digits starting from the lowest numbers being more frequent. If conflicts arise (i.e., a digit count doesn‚Äôt match its index), backtrack to adjust higher frequency of initial guesses.

2. **Re-evaluate Progress**: As digits aren't exceeding the 0-9 range, ensure the counts satisfy earlier assignments. The nearer you get to setting all digits, verify the accumulated sum is close to 10 without any contradiction.

3. **Identify Compatible Distribution**: The solution requires finding a balance where the allocation of digits aligns the structure heading both forwards (forming a candidate number) and backwards (verifying digit counts). Adjust errors progressively for any inconsistent or misassigned placements, correcting as you systematically resolve counts backward from misaligned assumptions.
